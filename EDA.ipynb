### DISTRIBUTION GRAPHS FOR THE WHOLE DATA SET

```python
# I will be getting the distribution of votes for all of the features for the whole data set
feature_df = df_1.drop(columns='party')
feature_df.head()
```

```python
# create list to loop through
feature_cols = list(feature_df.columns)
feature_cols
```

```python
df_1
```

```python
# GET DISTRIBUTIONS OF VOTES PER BILL
# Referred to https://stackoverflow.com/questions/64229894/how-to-fix-numpy-ndarray-object-has-no-attribute-get-figure-when-plotting-su
plt.tight_layout()
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(22, 22), sharey=True)

# unpack axes subplots to loop through
axe = axes.ravel()

for ix, col in enumerate(feature_cols):
    df_1[col].value_counts().sort_index().plot(kind='bar', ax=axe[ix])
    axe[ix].set_xlabel(f'Barchart for {col}', fontsize = 12)

plt.suptitle('Barchart of Votes per Bill', fontsize=20)
plt.show();
```

#### NOTES:
- Looking through these distributions of yes and no votes, I notice that some of the columns have very similar distributions. From the bottom, these include "crime", " 'aid-to-nicaraguan-contras', 'anti-satellite-test-ban', 'adoption-of-the-budget-resolution'. I am particularly interested in these columns because I suspect that adoption of the budget resolution is a very partisan bill.
- Another set of columns have a different kind of very similar distribution. From the bottom, these are 'superfund-right-to-sue', 'mx-missile', 'immigration', 'el-salvador-aid', 'water-project-cost-sharing'
- One column is unusual because it is the only one that had an overwhelming vote. This is the export administration act. There are relatively few '?' votes and no votes compared to yes votes. No other column is like that. I may want to drop this column.

### Distributions of votes for Republicans

```python
# Create a dataframe just containing Republican congressmemebers
df_repub = df_1[df_1['party'] == 'republican']
```

```python
df_repub
```

```python
df_repub = df_repub.drop(columns='party')
```

```python
# GET DISTRIBUTIONS OF VOTES PER BILL FOR REPUBLICANS
# Referred to https://stackoverflow.com/questions/64229894/how-to-fix-numpy-ndarray-object-has-no-attribute-get-figure-when-plotting-su

plt.tight_layout()
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(22, 22), sharey=True)

# unpack axes subplots to loop through
axe = axes.ravel()

for ix, col in enumerate(feature_cols):
    df_repub[col].value_counts().sort_index().plot(kind='bar', ax=axe[ix])
    axe[ix].set_xlabel(f'Barchart for {col}', fontsize = 12)

plt.suptitle('Barchart of Votes per Bill for Republicans', fontsize=20)
plt.show();
```

#### Notes

- Some of these votes are very highly correlated with being a Republican. These include a yes vote on the crime bill, a yes vote on the El Salvador aid bill, and a yes vote on the physician fee freeze bill

```python
# Create a dataframe with just Democrats
df_dem = df_1[df_1['party'] == 'democrat']
```

```python
df_dem = df_dem.drop(columns='party')
```

```python
df_dem
```

```python
# GET DISTRIBUTIONS OF VOTES PER BILL FOR DEMOCRATS
# Referred to https://stackoverflow.com/questions/64229894/how-to-fix-numpy-ndarray-object-has-no-attribute-get-figure-when-plotting-su

plt.tight_layout()
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(22, 22), sharey=True)

# unpack axes subplots to loop through
axe = axes.ravel()

for ix, col in enumerate(feature_cols):
    df_dem[col].value_counts().sort_index().plot(kind='bar', ax=axe[ix])
    axe[ix].set_xlabel(f'Barchart for {col}', fontsize = 12)

plt.suptitle('Barchart of Votes per Bill for Democrats', fontsize=20)
plt.show();
```

#### NOTES
- The El Salvador aid bill and the physician fee freeze bill stand out as partisan in this data and the Republican data above. Democrats usually voted no.
- Education spending and adoption of the budget resolution stand out as distinguishing Democrats from Republicans if you look at both sets of data

## Look for outliers among Republicans

```python
# put 1 in cells where Republicans voted yes on a bill and 0 in other cells
df_repub = df_repub.applymap(lambda x: int('1') if x =='y' else int('0'))
```

```python
df_repub
```

```python
df_repub.dtypes
```

```python
df_repub_sums = pd.DataFrame(df_repub.sum(axis=1), columns = ['yes_votes'])
```

```python
df_repub_sums
```

```python
sns.boxplot(x=df_repub_sums['yes_votes'])
```

**CONCLUSION**: The distribution of yes votes for Republicans is pretty even. There are not enough outliers to worry about.
